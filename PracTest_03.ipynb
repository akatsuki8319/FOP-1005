```python
import matplotlib.pyplot as plt
import numpy as np


def readdata(filename):
    data = []
    file1 = open(filename, 'r')
    lines = file1.readlines()          # read all lines of the file
    file1.close()
    for line in lines:
        # Use LIST COMPREHENSION to convert each value to int,
        # stripping whitespace/newline characters by splitting on comma
        splitline = [int(x) for x in line.strip().split(',')]
        data.append(splitline)
    return data


def convertArray(data):
    dataarray = np.array(data, dtype=int)   # convert list to numpy array
    return dataarray


def rowToImage(dataarray):
    data3D = []
    for d in dataarray:
        # Reshape each 1x784 row into a 28x28 numpy array and append to data3D
        image = d.reshape(28, 28)
        data3D.append(image)
    return data3D


def plotImages(data3D):
    plt.figure(1)
    plt.suptitle('Showing Fashion Dataset')

    plt.subplot(2, 4, 1)
    plt.imshow(data3D[0], cmap=plt.cm.gray)

    plt.subplot(2, 4, 2)
    plt.imshow(data3D[1], cmap='Blues')

    plt.subplot(2, 4, 3)
    plt.imshow(data3D[2], cmap='Greens')

    plt.subplot(2, 4, 4)
    plt.imshow(data3D[3], cmap='Reds')

    plt.subplot(2, 4, 5)
    plt.imshow(data3D[4], cmap='Purples')

    plt.subplot(2, 4, 6)
    plt.imshow(data3D[5], cmap='Oranges')

    plt.subplot(2, 4, 7)
    plt.imshow(data3D[6], cmap='viridis')

    plt.subplot(2, 4, 8)
    plt.imshow(data3D[7], cmap='plasma')

    # Save the final plot as FinalPlot.png
    plt.savefig('FinalPlot.png')
    plt.show()


def main():
    filename = input("Enter filename with extension ... ")

    # Call readdata function with the input file name
    data = readdata(filename)
    print('\nData read from file - represeting 1st image only\n')
    print(data[1])

    # Call convertArray function with the data list
    dataarray = convertArray(data)
    print('\nData converted to array - representing 1st image only\n')
    print(dataarray[1])

    # Call rowToImage function with the numpy array
    data3D = rowToImage(dataarray)
    print('\nData converted to 3D list - representing 1st image only\n')
    print(data3D[1])

    print('\nPlotting images\n')
    # Call plotImages function with the 3D list
    plotImages(data3D)

    print("End of code")


if __name__ == '__main__':
    main()
```
